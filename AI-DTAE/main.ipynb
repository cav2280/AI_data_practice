```python
# 1월부터 12월까지 합계 계산하여 "합계" 열로 추가
for month in range(1, 13):
    month_str = str(month)
    rain_df[month_str + '월'] = rain_df[month_str + '월'].astype(float)
    
rain_df['총 강수량'] = rain_df[[str(month) + '월' for month in range(1, 13)]].sum(axis=1)

# "이름"과 "합계" 열로 구성된 데이터프레임 출력
rain_sum_df = rain_df[['시점', '총 강수량']]
print(rain_sum_df)
```

```python
rain_sum_df
```

```python
rain_sum_df.rename(columns={'시점': '연도'}, inplace=True)
```

```python
rain_sum_df
```

```python
f_crops = crops_df.drop(['출저', '기준일자'], axis=1)
```

```python
f_crops
```

```python
df_result = pd.merge(f_crops, rain_sum_df, on='연도')
```

```python

df_result
```

```python
# 작물별 상관계수 계산 및 출력
all_correlations = {}

for crop in ['사과', '배', '매실', '단감', '무', '당근', '배추', '상추', '시금치', '미나리', '오이', '수박', '고추', '마늘', '양파', '대파', '감자', '고구마', '콩', '상황', '새송이', '표고', '딸기', '벼']:
    columns = [f'재배면적(ha)-{crop}', f'생산량(톤)-{crop}', f'농가수(가구)-{crop}', '총 강수량']
    df_crop = df_result[columns]
    correlation_matrix = df_crop.corr()
    all_correlations[crop] = correlation_matrix

# 결과 출력
for crop, correlation_matrix in all_correlations.items():
    print(f"\n{crop}:")
    print(correlation_matrix)
    
```

```python
# 데이터 프레임 생성 (예시를 위해 랜덤 데이터 사용)
data = {
    'Variable1': [1, 2, 3, 4, 5],
    'Variable2': [2, 4, 5, 4, 5],
}

df = pd.DataFrame(data)

# 상관관계 계산
correlation_matrix = df.corr()

# Heatmap으로 상관관계 그래프 그리기
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

plt.title('상관관계 그래프')
plt.show()
```

```python
# 상관관계 그래프 그리기
for crop, correlation_matrix in all_correlations.items():
    plt.figure(figsize=(6, 4))  # 그래프 크기 조정 (가로 6, 세로 4)
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

    plt.title(f'{crop} 상관관계 그래프')
    plt.show()
```
